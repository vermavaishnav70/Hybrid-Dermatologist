```python
from __future__ import annotations

import sys
from pathlib import Path

import cv2
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from IPython.display import display, Image

PROJECT_ROOT = Path.cwd()
if not (PROJECT_ROOT / 'src').exists():
    PROJECT_ROOT = PROJECT_ROOT.parent
if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))

from src.skin_analysis.pipeline import run_phase1_pipeline

plt.style.use('seaborn-v0_8-whitegrid')
pd.set_option('display.max_columns', 120)
pd.set_option('display.width', 140)
SEED = 42
```

```python
CANDIDATE_DATA_DIRS = [
    PROJECT_ROOT / 'dataset',
    PROJECT_ROOT / 'data' / 'raw' / 'Multi-Class Skin Condition Image Dataset (MSC-6)',
]
DATA_DIR = next((path for path in CANDIDATE_DATA_DIRS if path.exists()), CANDIDATE_DATA_DIRS[0])
OUTPUT_DIR = PROJECT_ROOT / 'outputs' / 'phase1_baseline'
IMAGE_SIZE = (224, 224)
TEST_SIZE = 0.2
RANDOM_STATE = 42
print('Using DATA_DIR:', DATA_DIR)
```

```python
metrics_path = OUTPUT_DIR / 'metrics_summary.csv'
if not metrics_path.exists():
    results = run_phase1_pipeline(
        data_dir=DATA_DIR,
        output_dir=OUTPUT_DIR,
        image_size=IMAGE_SIZE,
        test_size=TEST_SIZE,
        random_state=RANDOM_STATE,
    )
    metrics_df = results['metrics_summary']
else:
    metrics_df = pd.read_csv(metrics_path)

display(metrics_df)
```

```python
for model_name in ['svm', 'random_forest']:
    cm_path = OUTPUT_DIR / f'confusion_matrix_{model_name}.png'
    report_path = OUTPUT_DIR / f'classification_report_{model_name}.csv'
    print(f'--- {model_name.upper()} ---')
    if cm_path.exists():
        display(Image(filename=str(cm_path)))
    if report_path.exists():
        report_df = pd.read_csv(report_path, index_col=0)
        display(report_df.round(3))
```

```python
comparison = metrics_df.set_index('model')[['accuracy', 'precision', 'recall', 'f1_score']]
comparison.plot(kind='bar', figsize=(9, 5), ylim=(0.0, 1.0), colormap='viridis')
plt.title('Phase 1 Metric Comparison')
plt.ylabel('Score')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
